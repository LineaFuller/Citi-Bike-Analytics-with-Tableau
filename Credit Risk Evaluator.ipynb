```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Read CSV file into Pandas
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
train_df.drop(columns = ['Unnamed: 0',
 'index'], inplace= True)
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
test_df.drop(columns = ['Unnamed: 0',
 'index'], inplace= True)

train_df.head()

# Create data

# Separated Targets and Inputs
y_train = train_df["loan_status"]
X_train = train_df.drop(columns = "loan_status")
y_test = test_df["loan_status"]
X_test = test_df.drop(columns = "loan_status")

# Loan status: low risk = 0, high risk = 1
def convert_status (x):
    if x == "low_risk":
        return 0
    return 1

y_test = y_test.map(convert_status)
y_train = y_train.map(convert_status)

```

```python
# Convert categorical data to numeric and separate target feature for training data

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
train_cols = X_train.columns.tolist()
test_cols = X_test.columns.tolist()
                         
# Looking for column to drop from training data
for i in train_cols:
    if i not in test_cols:
        print(i)

# dummy_train.drop(columns="debt_settlement_flag_Y", inplace = True)

```

```python
# Predicition!

#I think my logistic regression model will perform better than my random forests classifier because /
# the number of observations is lesser than the number of features and will provide a discrete outcome. 
#
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")


# 
```

```python
# Train a Random Forest Classifier model and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

```python
# Outcome !

# Both of my outcomes were not great, but it appears my Random Forest Classifier model performed better
```

```python
# Scale the data

scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Prediction !

# I predict that my logistic regression model will perform better because scaling helps overcome /
# overfitting to get better results. 
```

```python
# Train the Logistic Regression model on the scaled data and print the model score

classifier = LogisticRegression()
classifier.fit(X_train_scaled, y_train)
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score

clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

```python
# Outcome! 

# My logistic regression model performed better 
```
